```python
import pandas as pd
import altair as alt

data = pd.read_csv("plotting_material.csv").iloc[ : , 1: ]
data['model'] = data['model']
data
#Max_num = data2["plotting_material.csv"].astype('int').max()
```

```python
alt.Chart(data).mark_bar().encode(
    x=alt.X('model', bin=alt.Bin(maxbins=40)),
    y='count()',
    color='model'
).properties(
        width=600,
        height=400
    )
```

```python
alt.Chart(data).mark_point(filled=True, size=400, opacity=0.5).encode(
    x='i',
    y='j',
).facet('model')

#alt.Chart(data).mark_square(filled=True, size=400, opacity=0.5).encode(
#    x='i',
#    y='j',
#).facet('model')
```

```python
data["a"] = pd.cut(data['i'].values, bins=40)
data["d"] = pd.cut(data['j'].values, bins=40)
maxi = data['model'].max()
new = data.groupby(["i","j"]).count()
new = new / maxi
new.to_csv("new.txt")
new
#B = data.query("Acceptor == 17 & Donor == 21" )
```

```python
col_list = ["i", "j","model"]
Normal_Data = pd.read_csv("new.txt", usecols=col_list)
Normal_Data = Normal_Data.rename({'model': 'Normalized Value'}, axis=1)

d = Normal_Data["Normalized Value"].sum()
Normal_Data
```

```python
graph = alt.Chart(Normal_Data).mark_rect().encode(
    alt.X('i', bin=alt.Bin(maxbins=40), scale=alt.Scale(domain=[0, 40]), title='Residue Index'),
    alt.Y('j', bin=alt.Bin(maxbins=40), title='Residue Index'),
    alt.Color('Normalized Value')).properties(
        width=600,
        height=400
    )
graph.interactive()
```
